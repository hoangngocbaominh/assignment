### How do you normalize the data?
   

```python
#Step1: First map the string values of diagnosis to integer. 
def mapping(data,feature):
    featureMap=dict()
    count=0
    for i in sorted(data[feature].unique(),reverse=True):
        featureMap[i]=count
        count=count+1
    data[feature]=data[feature].map(featureMap)
    return data
```

```python
data=mapping(data,"Title")
```

```python
#Step2: Drop the unrequired columns
data=data.drop(["Id","Company","Date","Location","Area","Classification","SubClassification","Requirement","FullDescription","JobType"],axis=1)
data.sample(5)
```

```python
#Step3: Normalizing dataframe
dataf=((data-data.min())/(data.max()-data.min()))*20
dataf.sample(10)
```
